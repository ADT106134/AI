```python
import folium
import pandas as pd
```

```python
df_earthquake = pd.read_csv(r'C:\Users\sino4\Desktop\Earthquake_2019.csV',encoding='big5')
```

```python
df_earthquake.head()
```

```python
m0 = folium.Map([23.5, 121], zoom_start = 7)
```

```python
m0
```

```python
folium.Marker(location = [df_earthquake['緯度'][0], df_earthquake['經度'][0]]).add_to(m0)
folium.Marker(location = [df_earthquake['緯度'][1], df_earthquake['經度'][1]]).add_to(m0)
folium.Marker(location = [df_earthquake['緯度'][2], df_earthquake['經度'][2]]).add_to(m0)
```

```python
m0
```

```python
m1 = folium.Map([23.5, 121], zoom_start = 7)
```

```python
for index, row in df_earthquake.iterrows():
    folium.Marker(location = [row['緯度'], row['經度']]).add_to(m1)
```

```python
m1
```

```python
m2 = folium.Map([23.5,121], zoom_start = 7)
```

```python
from folium.plugins import MarkerCluster
```

```python
marker_cluster = MarkerCluster().add_to(m2)
```

```python
for index, row in df_earthquake.iterrows():
    if row['規模'] >= 4.5:
        information = '地震時間:' + str(row['地震時間']) + '<br/>' + '規模:' + str(row['規模']) + '<br/>' + '深度:' + str(row['深度']) + '<br/>' + '位置:' + str(row['位置'])
        folium.Marker(location = [row['緯度'], row['經度']], popup = information).add_to(marker_cluster)
```
